```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('supermarket_sales.csv')
data.head()
```

```python
print(data.info())
```

```python
print(data.columns)
```

```python
data.columns=data.columns.str.lower().str.replace('_','')
print(data.columns)
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data['total_revenue'] =data['unit price']*data['quantity']
data.head()
```

```python
yangon_sales =data[data['city'] == 'Yangon']
yangon_sales.head()
```

```python
sales_by_product = data.groupby('product line')['sales'].sum().sort_values(ascending=False)
print(sales_by_product)
```

```python
avg_customer_spending = data.groupby('customer type')['sales'].mean()
print(avg_customer_spending)
```

```python
avg_customer_spending = data.groupby('customer type')['sales'].max()
print(avg_customer_spending)
```

```python
avg_customer_spending = data.groupby('customer type')['sales'].min()
print(avg_customer_spending)
```

```python
pivot_table=pd.pivot_table(
    data,
    values='sales',
    index='city',
    columns='product line',
    aggfunc='sum'
)
print("\nPivot table (City vs Product Line):")
print(pivot_table)
```

```python
branch_sales =data.groupby('branch')['sales'].sum().sort_values(ascending=False)
print(branch_sales)
```

```python
by_gender = data.groupby('gender')['sales'].sum().sort_values(ascending=False)
print(by_gender)
```

```python
payment_counts = data['payment'].value_counts()
print(payment_counts)
```

```python
sales_by_product.plot(kind='bar')
plt.title('Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```
