```python
ages = df['Age'] ## select single column
subset = df[['Name','Age','Fare']] #select multiple columns
old_passenger = df[df['Age'] > 30] ## Filter rows
adult_males = df[(df['Age'] > 18) & (df['Sex'] == 'male')]
```

```python
# Fill missing numerical values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill missing categorical with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop a column you don’t need
df.drop(columns=['Cabin'], inplace=True, errors='ignore')
```

```python
df['Sex'] = df['Sex'].map({   ##encoding map to numeric
    'male':0,
    'female':1
})
```

```python
# One-hot encoding for multiple categories
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] ##create new column

# Transform column (log, square root, etc.)
import numpy as np
df['Fare_log'] = np.log1p(df['Fare'])
```

```python
df.groupby('Pclass')['Fare'].mean() #Mean fare by class
```

```python
df.groupby('Sex')['Survived'].sum() #count survivors by gender
```

```python
df.groupby('Pclass').agg({  #multiple aggregation
    'Fare':'mean',
    'Age':'max'
})
```

```python
df.sort_values('Fare',ascending = False).head() #sort fare by descending order
```

```python
df.sort_values(['Pclass','Fare'], ascending=[True,False]).head() #sort by multiple columns
```
